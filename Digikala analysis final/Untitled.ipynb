# Load pars_ABSA data

## combine aspect and comments

```python
from xml.dom import minidom
import pandas as pd

xmldoc = minidom.parse('/home/admin/Desktop/Pars-ABSA/Pars-ABSA_xml.xml')
sentences = xmldoc.getElementsByTagName("sentence")
df = pd.DataFrame()
sentence_list = []
aspect_list = []
polarity_list = []

for sentence in sentences:
    sentence_text = sentence.getElementsByTagName("text")[0].firstChild.data
    aspect_terms = sentence.getElementsByTagName("aspectTerm")
    
    for i in range(len(aspect_terms)):
        sentence_list.append(sentence_text)
        aspect_list.append(aspect_terms[i].getAttribute('term'))
        polarity_list.append(aspect_terms[i].getAttribute('polarity'))

# Forming DataFrame
df['sentence'] = sentence_list
df['aspect'] = aspect_list
df['polarity'] = polarity_list
df = df.groupby('sentence', as_index=False).agg({'aspect': ', '.join, 'polarity': ', '.join})

from IPython.display import HTML
df.head(509)
# HTML(df.to_html(index=False))
# df.to_csv('Data.csv', index=False)
```

### llama3

```python
def En_Fa_seperator(row):
    my_list = re.split(':|,', row)

    persian = []
    english = []
    
    for item in my_list:
        item = item.replace("[","").replace("]", "").replace("\"", "").replace("\'", "")
        if 'Positive' in item or 'Negative' in item or 'Neutral' in item:
             english.append(item)
        else:
             persian.append(item)
    
    return persian,english


En_Fa_seperator(df3['En_aspect'][0])

# add new column to the dataframe 3 named En_sentiment
df3.insert(2, "En_sentiment", df3['En_aspect'], True)
    
for i in range(df3['En_aspect'].shape[0]):
    persian,english = En_Fa_seperator(df3['En_aspect'][i])
    df3.at[i, 'En_aspect'] = persian
    df3.at[i, 'En_sentiment'] = english

df3.to_csv('mostaghim(llama3).csv', index=False)
```
